```python
df = pd.read_csv('Support Ticket Customer Unresponsive Days.csv', encoding='latin1')
df
```

```python
#Convertimos en figuras geometricas
tickets_geo = gpd.GeoDataFrame(df, 
                                geometry=gpd.points_from_xy(df.Longitud, df.Latitud))

tickets_geo.head()
```

```python
#Revisamos que sí sea data geometrica
tickets_geo.dtypes
```

```python
#Ahora leemos el mapa 
#Primero el archivo Shp que descargamos de https://www2.census.gov/geo/tiger/TIGER_RD18/LAYER/STATE/
#NO OLVIDAR QUE NECESITAMOS EL ARCHIVO SHX, SI NO NO FUNCIONARÁ
#RUTA
shp_path = r'C:\Users\User\Maestria\Proyectos 1\Proyecto final\tl_rd22_us_state.shp'

# Leemos el shapefile
US_Map = gpd.read_file(shp_path)

# Muestra las primeras filas del GeoDataFrame
US_Map
```

```python
type(US_Map)
```

```python
#Esta en markdown porque aquí solo me da el número de Estado, sin saber el nombre
# Asignamos un estado a cada ticket con su geometria
tickets_geo['State'] = tickets_geo['geometry'].apply(lambda point: US_Map[US_Map.contains(point)]['geometry'].index[0] if not US_Map[US_Map.contains(point)].empty else None)

# Filtramos los tickets que no tienen un estado asignado
tickets_geo_filtered = tickets_geo.dropna(subset=['State'])

# Contamos el los tickets por estado
tickets_count_by_state = tickets_geo_filtered['State'].value_counts().reset_index()
tickets_count_by_state.columns = ['State', 'Ticket_Count']

# Fusionamos con el US_Map para obtener los nombres de los estados
tickets_count_by_state = pd.merge(tickets_count_by_state, US_Map[['geometry']], left_on='State', right_index=True, how='left')

# Mostramos el resultado final
#print(tickets_count_by_state)
```

```python
# Asignar nombres de directorio según el código de estado
state_directories = ['CALIFORNIA','ALABAMA', 'ARIZONA', 'ARKANSAS', 'CALIFORNIA', 'COLORADO', 'CONNECTICUT', 'DELAWARE', 'DISTRICT_OF_COLUMBIA', 'FLORIDA', 'GEORGIA', 'HAWAII', 'IDAHO', 'ILLINOIS', 'INDIANA', 'IOWA', 'KANSAS', 'PENNSILVANIA', 'LOUISIANA', 'MAINE', 'MARYLAND', 'MASSACHUSETTS', 'MICHIGAN', 'MINNESOTA', 'MISSISSIPPI', 'MISSOURI', 'MONTANA', 'NEBRASKA', 'NEVADA', 'NEW_HAMPSHIRE', 'NEW_JERSEY', 'NEW_MEXICO', 'NEW_YORK', 'NORTH_CAROLINA', 'NORTH_DAKOTA', 'OHIO', 'OKLAHOMA', 'OREGON', 'PENNSYLVANIA', 'RHODE_ISLAND', 'SOUTH_CAROLINA', 'SOUTH_DAKOTA', 'TENNESSEE', 'TEXAS', 'UTAH', 'VERMONT', 'VIRGINIA', 'WASHINGTON', 'WEST_VIRGINIA', 'WISCONSIN', 'WYOMING', 'VIRGINIA', 'GUAM', 'COMMONWEALTH_OF_THE_NORTHERN_MARIANA_ISLANDS', 'ARIZONA', 'UNITED_STATES_VIRGIN_ISLANDS', 'ALASKA',]

# Crear un diccionario para mapear los códigos de estado a nombres de directorio
state_code_to_directory = {state_code: state_directory for state_code, state_directory in enumerate(state_directories, start=1)}

# Asignar nombres de directorio según el código de estado
tickets_count_by_state['State_Directory'] = tickets_count_by_state['State'].map(state_code_to_directory)

# Fusionar con el US_Map para obtener las geometrías de los estados
tickets_count_by_state = pd.merge(tickets_count_by_state, US_Map[['geometry']], left_on='State', right_index=True, how='left')

# Mostrar el resultado final
print(tickets_count_by_state)
```
